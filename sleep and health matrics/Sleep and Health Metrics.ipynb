```python
rows, cols_layout = 4, 2
fig, axes = plt.subplots(rows, cols_layout, figsize=(12, 12))
axes = axes.flatten()

cols = [col for col in data.columns.to_list() if col != 'Sleep_Quality_Score']

for i, col in enumerate(cols):
    if i < len(axes):
        sns.scatterplot(x=col, y='Sleep_Quality_Score', data=data, ax=axes[i])
        axes[i].set_title(f'{col} vs Sleep_Quality_Score')

plt.tight_layout()
plt.show()
```
